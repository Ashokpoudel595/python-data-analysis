## USE OF SEABORN HEATMAP FUNCTION IN PYTH0N

```python
import pandas as pd
# import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Displaying random data matrix of 10x10 ranging from 0 to 100
data = np.random.randint(low=0, high=100, size=(10,10))
data
```

```python
# Hm-Heatmap
# using annot attribute
hm= sns.heatmap(data,annot=True,linewidth=1.5,linecolor="Red",cbar=True)
plt.show()
```

```python
# using linewidth and linecolor attribute
hm= sns.heatmap(data,linewidth=1.5,linecolor="Red")
plt.show()
```

```python
# enable or disable color bar attribute using cbar= True or False
hm= sns.heatmap(data,cbar=False)
```

```python
# using "xticklabels" and "yticklebels" to enable or disable labels accross row or columns 
hm= sns.heatmap(data,xticklabels=False,yticklabels=True)
```
